```python
from pathlib import Path
data_folder = Path("C:/Users/langzx/Desktop/github/EAmosm/data/notebook")
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import itertools
```

```python
demodata = pd.read_csv(data_folder/"demoorigin.csv")
list(demodata)
```

```python
nsize = len (demodata.index)
nsize
```

```python
obid = demodata ['ID']
Ob_sed = demodata ['SRed'] 
Ob_duck = demodata ['Duck']
Ob_cost = demodata ['Cost']

bcr_sc = [x/y for x, y in zip(Ob_sed , Ob_cost)]
bcr_dc = [x/y for x, y in zip(Ob_duck, Ob_cost)]


dict_new = {
    'ID': obid,
    'SRed': Ob_sed,
    'Duck':Ob_duck,
    'Cost':Ob_cost,
    'BCR_SC': bcr_sc,
    'BCR_DC': bcr_dc,
}

ks = pd.DataFrame(dict_new)
#ks
```

```python
ld = np.round(np.arange(0,1.1,0.1),1)
ld
```

```python
ldname = ["ld" + str(i) for i in ld]
ldname
```

```python
wht_ld = [[0] * nsize for j in range(len(ld))]
for j in range(len(ld)):
    ld_value = ld [j]
    wht_ld[j] = [ld_value * x + (1-ld_value) * y for x, y in zip(bcr_sc  , bcr_dc)]

ld_array = np.transpose(np.array(wht_ld))
```

```python
ld_df = pd.DataFrame(ld_array)
ld_df.columns = ldname
#ld_df
```

```python
df_dataset = pd.concat([ks.reset_index(drop=True), ld_df], axis=1)
#df_dataset
```

```python
ldrankname = ["rankbyld" + str(i) for i in ld]
ldrankname
```

```python
for i in range (len(ld)):
    df_dataset[ldrankname[i]] = df_dataset[ldname[i]].rank(ascending =False)

#df_dataset    
```

```python
top = np.round(np.arange(0.1,1,0.1),1)
```

```python
top_pct = top*100
topname = ["top" + str(int(i)) for i in top_pct]
topname
```

```python
ldtopname = [x + y for x in ldname for y in topname]
ldtopname
```

```python
nameindex = list (range(0,99,9))
nameindex
```

```python
for i in range (len (ld)):
    rankby = df_dataset[ldrankname[i]]
    for t in range (len (top)):
        topby = int(round(top[t] * nsize ))
        seed_ld = [0]*50 
        for j in range (nsize) :
            if rankby [j] <= topby:
                seed_ld [j] = 1
            else:
                seed_ld [j] = 0
        df_dataset[ldtopname[nameindex[i]+t]] = seed_ld
```

```python
df_dataset.to_csv(data_folder/"demodata_seeds.csv",index = False, sep=',', encoding='utf-8')
#df_dataset
```

```python
sed_sum = [0] * len(ldtopname)
duck_sum = [0] * len(ldtopname)
cost_sum = [0] * len(ldtopname)

for i in range (len(ldtopname)):
    sed_sum [i] = np.sum(df_dataset['SRed'] * df_dataset[ldtopname[i]])
    duck_sum [i] = np.sum(df_dataset['Duck'] * df_dataset[ldtopname[i]])
    cost_sum [i] = np.sum(df_dataset['Cost'] * df_dataset[ldtopname[i]])
```

```python
dict_front = {
    'SRed': sed_sum,
    'Duck': duck_sum ,
    'Cost': cost_sum 
}

front_df = pd.DataFrame(dict_front)
front_df
```

```python
front_df.to_csv(data_folder/"ldfront_seeds.csv",index = False, sep=',', encoding='utf-8')
```

```python

from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection='3d')
ax.scatter3D(sed_sum, duck_sum, cost_sum, cmap='viridis')
ax.set_xlabel('Sediment')
ax.set_ylabel('Duck')
ax.set_zlabel('Cost')
```
